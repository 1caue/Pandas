# Integração entre Python e Excel

```python
import pandas as pd
from openpyxl import workbook, load_workbook
tb = pd.read_excel(r'C:\Users\CAUÊ\Documents\Pandas\Excel\Products.xlsx')
```

```python
tb
```

```python
tb.loc[tb["Tipo"]=="Serviço", "Multiplicador Imposto"] = 1.5
tb
```

```python
tb['Preço Base Reais'] = tb['Multiplicador Imposto'] * tb['Preço Base Original']
tb
```

```python
planilha = load_workbook('Products.xlsx')
aba_atv = planilha.active
```

```python
for celula in aba_atv["C"]:
    if celula.value == 'Serviço':
        linha = celula.row
        aba_atv[f"D{linha}"] = 1.5

planilha.save("ProductsOpen.xlsx")
```
