# Apply e Map ↓

```python
# Criando o DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [16, 32, 53, 64],
    'C': [107, 145, 242, 132]},
    index=['Linha 1', 'Linha 2', 'Linha 3', 'Linha 4'])
```

```python
df
```

```python
# o .apply() é usado para aplicar uma função ao longo de um eixo de um DataFrame ou em valores de uma Series
def nossa_soma(series):
    return series.sum()
```

```python
# Somando valores com o .apply(), na horizontal ⭢
df['SOMA(A, B, C)'] = df.apply(nossa_soma, axis=1)
df
```

```python
# Somando valores na vertical ⬇ 
df.apply(nossa_soma, axis=0)
```

```python
# Calculando o valor Total das colunas e adicionando ao final do DF
df.loc['Total'] = df.apply(nossa_soma, axis=0)
df
```

```python
# Calculando a média das linhas e criando uma nova coluna ao DF
df['MEDIA(A, B, C)'] = df[['A', 'B', 'C']].apply(lambda series: series.mean(), axis=1)
df
```

```python
df['C * 2'] = df['C'].apply(lambda x: x * 2)
df
```

```python
df['A * 2'] = df['A'] * 2
df
```
