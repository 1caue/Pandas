```python
from IPython.display import display # Retornando a tabela 
import pandas as pd 

lista = pd.read_csv(r'C:\Users\CAUÊ\Documents\Pandas\Script\table.tsv', sep='\t')
```

```python
lista
```

```python
produto_view = lista['PRODUTO'] # Series Retornada referente á coluna NÃO É UMA CÓPIA mas sim uma REFERÊNCIA do DF
produto_view
```

```python
produto_bkp = lista['PRODUTO'].copy() # Agora sim essa serie é referente á uma coluna COPIADA
produto_bkp
```

```python
lista['PRODUTO'] = 'Combustível' # Atribuindo o valor "Combustível" a coluna "PRODUTO"
```

```python
lista.head() 
```

```python
produto_view
```

```python
produto_bkp
```

```python
nrows, ncols = lista.shape
nrows, ncols 
```

```python
new_products = [f'Produto {i}'for i in range(nrows) ]
len(new_products)
```

```python
# Quantidade de elementos de 'new_products' é igual ao número de linhas do DF
lista['PRODUTO'] = new_products
```

```python
lista
```

```python
print(produto_view) # O view muda pois é só uma referência
print('\n')
print('\n')
print(produto_bkp) # O bkp não, pois é um backup do DF
```

```python
# Voltando aos produtos Originais
lista['PRODUTO'] = produto_bkp # produto_bkp é uma Series, essa linha irá retornar ao valor Original do DF
```

```python
lista
```
