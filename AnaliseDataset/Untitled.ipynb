```python
import pandas as pd
from IPython.display import display

data1 = pd.read_csv(r'C:\Users\CAUÊ\Documents\Pandas\AnaliseDataset\campeonato-brasileiro-cartoes.csv')
data2 = pd.read_csv(r'C:\Users\CAUÊ\Documents\Pandas\AnaliseDataset\campeonato-brasileiro-gols.csv')

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

main_df = df1.set_index('partida_id').join(df2.set_index('clube'), rsuffix='_tab2')

display(main_df)
```

```python
jogadores = pd.DataFrame({'Jogador ': ['Pedro', 'João', 'Vitor', 'Ryan'], 
                          'Gols': [3, 2, 6, 1], 
                          'Cartão':['Amarelo', 'Amarelo', 'Vermelho', 'Nenhum']})
display(jogadores)
```

```python
time = pd.DataFrame({'Time': ['Flamengo', 'Flamengo', 'Fluminense', 'Vasco'], 
                     'Gols na Liga': [4, 5, 8, 1],
                     'Partidas': [4, 5, 5, 4]})

display(time)
```

```python
tot = jogadores.join(time) # Unindo os dataframes utilizando o .join()
display(tot)
```

```python
import matplotlib.pyplot as plt # Formatando DataFrame e mostrando na tela 

plt.figure(figsize=(10, 1))
plt.table(cellText=tot.values, colLabels=tot.columns, cellLoc='left', loc='center')
plt.axis('off')
plt.show()
```

```python
tot.plot(kind='line', grid=True, title="Gráfico", ylabel="Y", xlabel="X") # Criando gráfico
```
